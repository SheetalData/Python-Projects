```python
fig = plt.figure()
```

```python
fig = plt.figure()
ax =  fig.add_subplot(1,2,1)#AXES=Subplots
ax.plot(x,y1)
```

```python
fig = plt.figure()
ax =  fig.add_subplot(1,2,1)#AXES=Subplots
ax.plot(x,y1)
plt.title('Title1')

ax =  fig.add_subplot(1,2,2)#AXES=Subplots
ax.plot(x,y2)
plt.title('Title2')
```

###### Another way to create fig axes

```python
fig,ax= plt.subplots(1,2)
ax[0].plot(x,y1)
ax[1].plot(x,y2)
```
